# Make median_house_value binary

```python
mean = data['median_house_value'].mean()
```

```python
data['above_average'] = np.where(data['median_house_value'] > mean , 1 , 0)
```

```python
data = data.drop('median_house_value', axis=1)
```

# Split the data

```python
from sklearn.model_selection import train_test_split 
# We importing train_test_split from sklearn
```

```python
data_full_train, data_test = train_test_split(data, test_size=0.2,random_state=1)
# We are now spliting the data_test data from original data, only 20%  
```

```python
data_train, data_validation = train_test_split(data_full_train, test_size=0.25, random_state=1)
# We are now spliting the data into data_train and data_validation,the data_validation is only 25% 
# of the data_full_train 
```

```python
y_train = data_train.above_average.values
y_validation = data_validation.above_average.values
y_test = data_test.above_average.values
# Here we are creating the y variables, for the trainings set, the validation set and test set.
```

```python
data_train = data_train.reset_index(drop=True)
data_validation = data_validation.reset_index(drop=True)
data_test = data_test.reset_index(drop=True)
# Here we are reseting the Index of the different data set.
```

```python
del data_train['above_average']
del data_validation['above_average']
del data_test['above_average']
# Delete the median_house_value variable from the data_train, data_validation, data_test
```

```python
from sklearn.metrics import mutual_info_score
# importing the mutual info score
```

```python
categorical = ['longitude', 'latitude', 'housing_median_age', 'total_rooms',
        'total_bedrooms', 'population', 'households', 'median_income', 'ocean_proximity', 'rooms_per_household',
        'bedrooms_per_room', 'population_per_household']
```

```python
def mutual_info_median_house_value_score(series):
    return mutual_info_score(series, data_full_train.above_average.values)
# making a function for the mutau information. 
```

```python
mutual_outcome = data_full_train[categorical].apply(mutual_info_median_house_value_score)
mutual_outcome.sort_values(ascending=True)
# looking for the one with the highest rate of mutual information 
```
